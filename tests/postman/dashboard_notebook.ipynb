```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import pandas as pd
```

## Dados de Exemplo

```python
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Value1': [10, 15, 13, 17],
    'Value2': [16, 5, 11, 9]
})
```

## Inicialize o Aplicativo Dash

```python
app = dash.Dash(__name__)
```

## Layout do Dashboard

```python
app.layout = html.Div(children=[
    html.H1(children='Dashboard Exemplo'),

    html.Div(children='''
        Dashboard com vários gráficos usando Dash e Plotly.
    '''),

    dcc.Graph(
        id='bar-graph',
        figure={
            'data': [
                go.Bar(
                    x=df['Category'],
                    y=df['Value1'],
                    name='Value1'
                ),
                go.Bar(
                    x=df['Category'],
                    y=df['Value2'],
                    name='Value2'
                )
            ],
            'layout': {
                'title': 'Bar Graph'
            }
        }
    ),

    dcc.Graph(
        id='line-graph',
        figure={
            'data': [
                go.Scatter(
                    x=df['Category'],
                    y=df['Value1'],
                    mode='lines+markers',
                    name='Value1'
                ),
                go.Scatter(
                    x=df['Category'],
                    y=df['Value2'],
                    mode='lines+markers',
                    name='Value2'
                )
            ],
            'layout': {
                'title': 'Line Graph'
            }
        }
    )
])
```

## Execute o Servidor

```python
if __name__ == '__main__':
    app.run_server(debug=True)
```
